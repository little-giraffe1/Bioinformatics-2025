**Множественное выравнивание белковых последовательностей**

Для выравнивания были использованы последовательности белка PDS (phytoenedesaturase) арабидопсиса, нарцисса, риса и красного перца. PDS синтезирует один из предшественников каротиноидов. 

Выравнивание производилось инструментом MUSCLE. Приведем полученное выравнивание.

(Приведем сначала в формате Clustal, чтобы посмотреть на результат) 





В формате Newick выравнивание представляется в виде дерева:

(

(

sp|A2XDA1|PDS_ORYSI:0.10365,

sp|Q40406|PDS_NARPS:0.08967)

:0.02176,

sp|P80093|PDS_CAPAN:0.10761,

sp|Q07356|PDS_ARATH:0.08989);

С помощью инструмента TreeViewer построим дерево.



Видим, что получилось довольно осмысленное дерево: рис и нарцисс оказались в одной ветви, они относятся к однодольным. К этой ветви сестринская группа - красный перец, Piperaceae - это базальная и довольно примитивная группа двудольных. Арабидопсис выделился в сестринскую группу по отношению ко всем остальным, так как это достаточно продвинутое двудольное.
