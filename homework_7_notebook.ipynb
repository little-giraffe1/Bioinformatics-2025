1. Последовательности S-белка SARS CoV2

Для данного задания было выбрано 4 последовательности S-белка разных вариантов коронавируса SARS CoV2: уханьский вариант (wuhan), дельту, омикрон и лямбда, а также последовательность 10825 из семинара. 

Для построения дерева использовалась программа MEGA. Был произведен подбор модели (вкладка Models), наилучшей оказалась модель JTT (матрица замен для белков, подобная РАМ, но построенная по более новым базам данных). 

Полученное дерево приведено в виде кладограммы и в виде филограммы (длины ветвей соответствуют относительной разнице между последовательностями, которая также указана численно).

Видно, что наиболее похожи последовательности S-белков вариантов wuhan и lambda. Сестринской группой к ним является omicron, а затем delta (причем они отделились примерно в один момент, так как на филограмме этот узел практически неразрешен). Последовательность 10825 выделилась отдельно от остальных. 

Видно, что последовательность S-белка delta относительно сильно отличается от остальных (примерно настолько же, как и 10825, которая бралась в качестве аутгруппы).

В данном задании использовались белковые последовательности фосфоенолпируваткарбоксилазы для растений разных семейств: проса, кукурузы (сем. Злаковые), кувшинки (сем. Кувшинковые), тополя (сем. Ивовые), киноа (сем. Маревые), китайской розы (сем. Розовые), ипомеи (сем. Пасленовые), хлопка (сем. Мальвовые), клещевины (сем. Молочайные) и пекана (сем. Ореховые). Фосфоенолпируваткарбоксилаза присутствует у всех растений и катализирует карбоксилирование ФЕП, что является одной из анаплеротических реакций цикла Кребса, а также участвует в С4-фотосинтезе у некоторых растений (просо, кукуруза), поэтому можно предположить, что она достаточно консервативна, чтобы ее можно было использовать для построения филогенетического дерева.  

Последовательности для ортологов были взяты из базы данных NCBI Protein, выравнивание и построение дерева производилось с помощью MEGA. 

Было построено дерево Maximum Likelihood с моделью JTT, которая была выбрана алгоритмом подбора как лучшая.
