Видим, что исходные риды содержат адаптеры и в целом низкое среднее качество (warning для R1 и fail для R2). Образец R1 содержит некие повторяющиеся последовательности и имеет проблемы с GC-составом. Образец R2 имеет проблемы с Per Base Sequence Content, что может говорить о не очень хорошем качестве секвенирования.

Сделам тримминг с помощью TRIMMOMATIC. Необходимо обрезать адаптеры (ILLUMINACLIP), также отрежем части с низким качеством в начале и в конце (LEADING:20 и TRAILING:20), применим SLIDINGWINDOW, чтобы увеличить среднее качество рида по длине; в конце применим MINLEN, чтобы выбросить слишком короткие риды.

Сделаем fastqc-отчет ридов после тримминга и сравним с ридами до тримминга.



Видим, что адаптеры были успешно обрезаны. Для R1 удалось улучшить среднее качество ридов, также все риды, содержащие повторяющиеся последовательности, оказались в файле с неспаренными ридами. Проблема с GC-составом сохраняется. Для R2 среднее качество ридов и содержание нуклеотидов привести к хорошему не удалось. Вообще файл R2_unpaired имеет многочисленные проблемы - с качеством, средним по позиции и по последовательности, и содержанием нуклеотидов. Также у всех четырех полученных файлов появились проблемы со средней длиной ридов.
