В данной работе выполнялся поиск HMM-мотива белка 14-3-3 в протеомах арабидопсиса Arabidopsis thaliana и подсолнечника Helianthus annuus. Для поиска был выбран hmmsearch, так как он выполняет поиск HMM-мотива среди белковых последовательностей. Вывод был оформлен в формате .tbl, для каждой найденной последовательности указан e-value, покрытие, скор, название и описание. 

Далее для каждой найденной последовательности приведено выравнивание на мотив 14-3-3.

Для каждого организма первые две последовательности (с минимальным e-value) были скачаны из базы данных Protein (NCBI), было произведено множественное выравнивание в программе MEGA. В выравнивание также была добавлена последовательность мотива 14-3-3. 

Видно, что качество выравнивания довольно хорошее, но все равно довольно много полиморфизмов. Скорее всего, это связано с тем, что hmmersearch может находить более далекие от мотива последовательности, чем, напрмиер, BLAST.
